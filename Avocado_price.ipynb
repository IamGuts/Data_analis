```python
# Расчет средней цены
average_price_conventional = conv_df['AveragePrice'].mean()
average_price_organic = organic_df['AveragePrice'].mean()
```

```python
# Анализ объема продаж
total_volume_conventional = conv_df['Total Volume'].sum()
total_volume_organic = organic_df['Total Volume'].sum()
```

```python
labels = ['Конвенциональные', 'Органические']
average_prices = [average_price_conventional, average_price_organic]
volumes = [total_volume_conventional, total_volume_organic]

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# График для средней цены
ax[0].bar(labels, average_prices, color=['orange', 'green'])
ax[0].set_title('Средняя цена авокадо')
ax[0].set_ylabel('Цена ($)')
ax[0].set_ylim(0, max(average_prices) * 1.2)

# График для объема продаж
ax[1].bar(labels, volumes, color=['orange', 'green'])
ax[1].set_title('Объём продаж авокадо')
ax[1].set_ylabel('Объём продаж')
ax[1].set_ylim(0, max(volumes) * 1.2)

plt.tight_layout()
```

```python
# Группировка данных по годам и регионам с подсчётом общего объёма продаж
sales_volume = avocado_df.groupby(['year', 'region'])['Total Volume'].sum().reset_index()
sales_volume
```

```python

# Визуализация данных
plt.figure(figsize=(10, 6))
for region in sales_volume['region'].unique():
    plt_data = sales_volume[sales_volume['region'] == region]
    plt.plot(plt_data['year'], plt_data['Total Volume'], marker='o', label=region)

plt.title('Общий объём продаж авокадо по годам и регионам')
#plt.xlabel('Год')
#plt.ylabel('Общий объём продаж')
plt.xticks(sales_volume['year'].unique())
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
```

```python
# Вычисляем общий объем продаж
avocado_df['Total Volume'] = avocado_df['4046'] + avocado_df['4225'] + avocado_df['4770']
```

```python
# Функция для расчета вклада каждого PLU в общий объем продаж
def calculate_contribution(data, plu_column):
    contribution = data[plu_column].sum() / data['Total Volume'].sum() * 100
    return contribution
```

```python
# Рассчитываем вклад для каждого PLU
contribution_4046 = calculate_contribution(avocado_df, '4046')
contribution_4225 = calculate_contribution(avocado_df, '4225')
contribution_4770 = calculate_contribution(avocado_df, '4770')

# Выводим результаты
print(f'Вклад PLU 4046 в общий объем продаж: {contribution_4046:.2f}%')
print(f'Вклад PLU 4225 в общий объем продаж: {contribution_4225:.2f}%')
print(f'Вклад PLU 4770 в общий объем продаж: {contribution_4770:.2f}%')
```

```python
plu_labels = ['PLU 4046', 'PLU 4225', 'PLU 4770']
contributions = [contribution_4046, contribution_4225, contribution_4770]


plt.figure(figsize=(8, 5))
plt.bar(plu_labels, contributions, color=['blue', 'orange', 'green'])

#plt.title('Вклад PLU в общий объем продаж')
#plt.xlabel('PLU')
#plt.ylabel('Вклад (%)')


for i, v in enumerate(contributions):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center')

plt.ylim(0, max(contributions) + 10)  # Увеличиваем предел по оси Y для удобства
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
regional_summary = avocado_df.groupby('region').agg(
    AveragePrice=('AveragePrice', 'mean'),
    TotalVolume=('Total Volume', 'sum')
).reset_index()
regional_summary.info()
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(data=regional_summary, x='AveragePrice', y='region', palette='viridis')
#plt.title('Средняя цена авокадо по регионам')
#plt.xlabel('Средняя цена (доллары)')
#plt.ylabel('Регион')
plt.grid(axis='x')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=regional_summary, x='TotalVolume', y='region', palette='magma')
#plt.title('Общее количество продаж авокадо по регионам')
#plt.xlabel('Общее количество продаж')
#plt.ylabel('Регион')
plt.grid(axis='x')
plt.tight_layout()
```
