### Задание 1

Из списка выбрать индексы тех элементов, которые в сумме дают результат. Использовать одни и теже элементы дважды нельзя. Порядок вывода значения не имеет.

***Пример 1***

Входящий список: nums = [2,7,11,15], результат - 9

Выход: [0,1]

Объяснение: поскольку 2 + 7 = 9, а позиция элементов 0 и 1

***Пример 2:***

Входящий список: nums = [3,2,4], результат = 6

Выход: [1,2]

Объяснение: поскольку 2 + 4 = 6, а позиция элементов 1 и 2


```python
nums_1 = [2,7,11,15]
nums_2 = [3,2,4]
def funk(data):
    for index_1 in range(len(data)):
        for index_2 in range(len(data)):
            if index_1 != index_2:
                print(f'Обьяснение: по скольку {data[index_1]} + {data[index_2]} = {data[index_1] + data[index_2]} , а позиция элементов {index_1}, {index_2}')


print(f'Для списка {nums_1}')
print(funk(nums_1))
print(f'Для списка {nums_2}')
print(funk(nums_2))
```

### Задание 2

В данной строке, включающей слова и пробелы, подсчитать количество символов в последнем слове. Слово - максимальная подстрока, внутри которой нет пробелов


***Пример 1:***

Входящая строка: s = "Hello World"

Выход: 5

Объяснение: Последнее слово "World" и его длина равна 5.


***Пример 2:***

Входящая строка: s = "   fly me   to   the moon  "

Выход: 4

Объяснение: Последнее слово "moon" и его длина равна 4.

```python
### Ваш код здесь
st_1 = ' Hello World'
st_2 = "fly me to the moon"

def len_word(string):
    st = string.split()[-1]
    return f'Объяснение: Последнее слово {st} и его длина равна: {len(st)} '

print(len_word(st_1))
print(len_word(st_2))
```

### Задание 3


Дано целое число INT. Напишите функцию, которая возвращает TRUE, если число ПАЛИНДРОМ(одинаково как сначала в конец, так и наоборот), и FALSE в противном случае.

***Пример 1:***

Вход: 121

Выход: TRUE

***Пример 2:***

Вход: 345

Выход: FALSE



```python
from random import randint

random_list = [randint(9,1000) for i in range(10)]

def polindrom(data):
    for n in data:
        if str(n) == str(n)[::-1]:
            print(f'True ({n})')

        else:
            print('False')
print(f'Сам Список {random_list}')
polindrom(random_list)
```
