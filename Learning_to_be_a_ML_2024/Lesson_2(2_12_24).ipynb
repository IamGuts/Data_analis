## Different types of graphs

### Bar plot

```python
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 12), xticks=np.arange(1, 8),
       ylim=(0, 12), yticks=np.arange(1, 8))

plt.show()
```

### Stack plot

```python
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])


fig, ax = plt.subplots()

ax.stackplot(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

#### Histogramm

```python
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)


fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

#### Pie plot

```python
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))


fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

#### 3D scatter plot

```python
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

## Seaborn


```python
import seaborn as sns 
df = sns.load_dataset('penguins')
```

```python
df.head()
```

### Simple scatterplot

```python
sns.set_style('whitegrid')
```

```python
sns.scatterplot(data=df, x='bill_length_mm', y='flipper_length_mm')
plt.show()
```

### Relplot. Kind - scatter

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter')
plt.show()
```

#### Adding column (col)

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter',col='sex')
plt.show()
```

#### Adding hue by categorical column (species)

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter', hue='species', col='sex')
```

#### Using size instead of hue

```python
sns.relplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='scatter', size='body_mass_g', hue='sex')
```

```python
sns.boxplot(data=df, x='bill_length_mm')
```
